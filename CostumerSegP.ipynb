<h3><b>2.6 Outlier Detection</b></h3>

```python
if numerical_cols:
    print("OUTLIER DETECTION USING IQR METHOD:")
    for col in numerical_cols:
        data = df[col].dropna()
        if len(data) > 0:
            Q1 = data.quantile(0.25)
            Q3 = data.quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            outliers = data[(data < lower_bound) | (data > upper_bound)]
            outlier_count = len(outliers)
            outlier_percent = (outlier_count / len(data)) * 100
            print(f"{col}: {outlier_count} outliers ({outlier_percent:.2f}%)")
```

<h3><b> 3.2 Missing Value Imputation</h3>

```python
# Create a copy for cleaning
df_clean = df.copy()

# 1. Drop rows where birthdate is missing
df_clean.dropna(subset=['customer_birthdate'], inplace=True)

# 2. Impute with 0 (Logic-based)
zero_cols = ['kids_home', 'teens_home', 'number_complaints']
df_clean[zero_cols] = df_clean[zero_cols].fillna(0)

# 3. Impute with Median (Robust to outliers)
median_cols = [
    'lifetime_spend_electronics', 'lifetime_spend_vegetables', 
    'lifetime_spend_meat', 'lifetime_spend_fish', 
    'lifetime_spend_videogames', 'lifetime_spend_petfood',
    'lifetime_spend_alcohol_drinks', 'lifetime_spend_hygiene',
    'typical_hour', 'percentage_of_products_bought_promotion',
    'distinct_stores_visited'
]
for col in median_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].median())

# 4. Impute Loyalty Card
df_clean['loyalty_card_number'] = df_clean['loyalty_card_number'].fillna('No Card')
```

<h3><b> 3.3 Handling outliers</h3>

```python
# Create Age (You already have this, but make sure to run it after fixing dates)
df_clean['age'] = 2026 - pd.to_datetime(df_clean['customer_birthdate'], errors='coerce').dt.year

# Check for unrealistic ages
print(f"Oldest customer: {df_clean['age'].max()}")
print(f"Youngest customer: {df_clean['age'].min()}")

# Logic: Remove or cap ages that don't make sense (e.g., older than 100 or younger than 18)
df_clean = df_clean[(df_clean['age'] <= 100) & (df_clean['age'] >= 18)]

print(f"Dataframe size after age filtering: {df_clean.shape[0]} rows")
```

```python
# Identify numerical columns for outlier treatment
cols_to_cap = [
    'age', 'kids_home', 'teens_home', 'number_complaints', 
    'distinct_stores_visited', 'lifetime_spend_groceries', 
    'lifetime_spend_electronics', 'lifetime_spend_vegetables', 
    'lifetime_spend_nonalcohol_drinks', 'lifetime_spend_alcohol_drinks', 
    'lifetime_spend_meat', 'lifetime_spend_fish', 'lifetime_spend_hygiene', 
    'lifetime_spend_videogames', 'lifetime_spend_petfood', 
    'percentage_of_products_bought_promotion'
]

for col in cols_to_cap:
    if col in df_clean.columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Capping the values at the bounds
        df_clean[col] = df_clean[col].clip(lower=lower_bound, upper=upper_bound)

print("Outliers have been capped using the IQR method.")
```

<h3><b> 4.Feature Engineering</h3>

```python
# A. Create Age from Birthdate (using the project year 2026)
df_clean['age'] = 2026 - pd.to_datetime(df_clean['customer_birthdate']).dt.year

# B. Create Tenure (Years as a customer)
df_clean['tenure'] = 2026 - df_clean['year_first_transaction']

# C. Aggregate Spending: Total Lifetime Spend [cite: 39-46]
spend_cols = [col for col in df_clean.columns if 'lifetime_spend' in col]
df_clean['total_spend'] = df_clean[spend_cols].sum(axis=1)

# D. Household Dynamics: Total Children
df_clean['total_children'] = df_clean['kids_home'] + df_clean['teens_home']

# E. Loyalty Status: Binary Flag
df_clean['is_loyalty_member'] = df_clean['loyalty_card_number'].apply(lambda x: 0 if x == 'No Card' else 1)
```
