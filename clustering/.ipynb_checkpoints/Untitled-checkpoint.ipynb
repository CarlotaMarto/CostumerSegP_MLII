```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.decomposition import PCA
```

```python
path = "/Users/franciscateixeira/Documents/ML/CostumerSegP_MLII/datasets/customer_info_clean.csv"

df = pd.read_csv(path, index_col="customer_id")

display(df.head())
print(df.shape)
display(df.info())
```

```python
print("Missing values:", df.isna().sum().sum())
print("Duplicated rows:", df.duplicated().sum())

display(df.describe().T)
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
dispersion = []

for k in range(1, 21):
    kmeans = KMeans(n_clusters=k, random_state=0, n_init=20)
    kmeans.fit(scaled_data)
    dispersion.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 21), dispersion, marker="o")
plt.xlabel("Number of clusters")
plt.ylabel("Dispersion / Inertia")
plt.title("Elbow Curve - KMeans")
plt.grid(True)
plt.show()
```

```python
plt.plot(range(1, 21), dispersion, marker="o")
plt.xlabel("Number of clusters")
plt.ylabel("Dispersion / Inertia")
plt.title("Elbow Curve - KMeans")

plt.vlines(4, min(dispersion), max(dispersion), color="red", linestyle="--")
plt.vlines(6, min(dispersion), max(dispersion), color="red", linestyle="--")

plt.grid(True)
plt.show()
```

```python
silhouette_scores = []

for k in range(2, 21):
    kmeans = KMeans(n_clusters=k, random_state=0, n_init=20)
    labels = kmeans.fit_predict(scaled_data)
    silhouette_scores.append(silhouette_score(scaled_data, labels))
```

```python
plt.plot(range(2, 21), silhouette_scores, marker="o")
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score - KMeans")
plt.grid(True)
plt.show()
```
