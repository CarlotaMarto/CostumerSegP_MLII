# Customer Segmentation - Clustering

Continuation after the EDA notebook `eda_costumer_info.ipynb`.

This notebook uses the cleaned file exported by the EDA phase: `customer_info_clean.csv`.

```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from clustering_utils import (
    RANDOM_STATE,
    DEFAULT_EXCLUDED_COLS,
    read_customer_info_clean,
    check_clustering_inputs,
    prepare_features,
    evaluate_kmeans_range,
    run_kmeans,
    cluster_profile,
    relative_cluster_profile,
    visualize_pca,
    visualize_umap,
    silhouette_plot,
    hierarchical_on_centroids,
)
```

## 1. Load the cleaned dataset from EDA

```python
# This file is created at the end of your EDA notebook:
# ci_clean.to_csv(.../datasets/customer_info_clean.csv, index=True)

customer_info_clean = read_customer_info_clean("../datasets/customer_info_clean.csv")

print("Dataset shape:", customer_info_clean.shape)
display(customer_info_clean.head())
```

## 2. Check columns used by KMeans

```python
excluded_cols = DEFAULT_EXCLUDED_COLS

input_check = check_clustering_inputs(customer_info_clean, exclude_cols=excluded_cols)

print("Columns used by KMeans:")
print(input_check["features_used"])

print("\nExcluded columns found in the dataset:")
print(input_check["excluded_existing"])

print("\nNon-numeric columns not used by KMeans:")
print(input_check["non_numeric_not_used"])
```

## 3. Choose the number of clusters with Elbow and Silhouette

```python
kmeans_results = evaluate_kmeans_range(
    customer_info_clean,
    min_k=2,
    max_k=10,
    exclude_cols=excluded_cols,
    random_state=RANDOM_STATE
)

display(kmeans_results)
```

## 4. Fit final KMeans model

Change `FINAL_K` after checking the elbow and silhouette results.

```python
FINAL_K = 6

labels, centroids, silhouette, kmeans_model, feature_cols = run_kmeans(
    customer_info_clean,
    k=FINAL_K,
    exclude_cols=excluded_cols,
    random_state=RANDOM_STATE
)

customer_segments = customer_info_clean.copy()
customer_segments["kmeans_cluster"] = labels

print(f"Final KMeans silhouette score: {silhouette:.4f}")

print("\nCluster sizes:")
display(customer_segments["kmeans_cluster"].value_counts().sort_index())

print("\nCentroids:")
display(centroids)
```

## 5. Visualize clusters

```python
visualize_pca(
    customer_segments,
    labels=customer_segments["kmeans_cluster"],
    exclude_cols=excluded_cols,
    title="PCA - KMeans clusters"
)
```

```python
# Optional: requires umap-learn
# If needed, install it first with: pip install umap-learn

visualize_umap(
    customer_segments,
    labels=customer_segments["kmeans_cluster"],
    exclude_cols=excluded_cols,
    title="UMAP - KMeans clusters"
)
```

## 6. Silhouette plot

```python
X, feature_cols = prepare_features(customer_segments, exclude_cols=excluded_cols)

silhouette_plot(
    X,
    customer_segments["kmeans_cluster"].values,
    title="Silhouette Plot - KMeans"
)
```

## 7. Cluster profiling

```python
profile = cluster_profile(
    customer_segments,
    cluster_col="kmeans_cluster",
    exclude_cols=excluded_cols
)

display(profile)
```

```python
relative_profile = relative_cluster_profile(profile)

display(relative_profile)
```

## 8. Optional: Hierarchical clustering over KMeans centroids

```python
N_MACRO_CLUSTERS = 6

Z, macro_labels, centroid_to_macro = hierarchical_on_centroids(
    centroids,
    n_macro_clusters=N_MACRO_CLUSTERS,
    plot=True
)

customer_segments["macro_cluster"] = (
    customer_segments["kmeans_cluster"]
    .map(centroid_to_macro)
)

print("Macro cluster sizes:")
display(customer_segments["macro_cluster"].value_counts().sort_index())
```

## 9. Save final clustered dataset

```python
output_path = "../datasets/customer_segments_final.csv"

customer_segments.to_csv(output_path, index=True)

print(f"Saved final clustered dataset to: {output_path}")
display(customer_segments.head())
```
